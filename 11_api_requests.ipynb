```python
import os
import pandas as pd
from dotenv import load_dotenv

load_dotenv()
```

```python
client_id = os.getenv('SPOTIFY_CLIENT1')
client_secret = os.getenv('SPOTIFY_SECRET1')
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import matplotlib.pyplot as plt

sp = spotipy.Spotify(
    auth_manager=SpotifyClientCredentials(
        client_id=client_id, client_secret=client_secret
        )
        )
```

```python
result = sp.search(q='coldplay', limit=50)
res = []

for idx, track in enumerate(result['tracks']['items']):
    res.append({
        'track_name': track['name'],
        'popularity': track['popularity'],
        'duration_ms': track['duration_ms'],
    })

```

```python
df_raw = pd.DataFrame(res)
df = df_raw.copy()
df
```

```python
plt.scatter(df['duration_ms'], df['popularity'])
plt.show()
```
